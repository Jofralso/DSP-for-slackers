```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple discrete-time signal
n = np.arange(0, 10)
x = np.sin(0.2 * np.pi * n)

# Plot the signal
plt.stem(n, x)
plt.title('Discrete-Time Signal')
plt.xlabel('n (Sample Index)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Simulate sampling and reconstruction
t_continuous = np.linspace(0, 10, 1000)
x_continuous = np.sin(0.2 * np.pi * t_continuous)

t_sampled = np.arange(0, 10, 1)
x_sampled = np.sin(0.2 * np.pi * t_sampled)

# Reconstruct the continuous signal (ideal)
x_reconstructed = np.interp(t_continuous, t_sampled, x_sampled)

# Plot the original and reconstructed signals
plt.plot(t_continuous, x_continuous, label='Continuous Signal')
plt.stem(t_sampled, x_sampled, linefmt='r--', markerfmt='ro', basefmt='r-')
plt.plot(t_continuous, x_reconstructed, label='Reconstructed Signal (Ideal)')
plt.title('Sampling and Reconstruction')
plt.xlabel('t (Time)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients of a difference equation
b = [1, 2, 1]
a = [1, -0.6, 0.1]

# Generate an input signal
x = np.sin(0.2 * np.pi * np.arange(0, 10))

# Apply the difference equation
y = np.zeros_like(x)
for n in range(len(x)):
    y[n] = sum(b[k] * x[n - k] for k in range(len(b))) - sum(a[k] * y[n - k] for k in range(1, len(a)))

# Plot the input and output signals
plt.stem(x, label='Input Signal')
plt.stem(y, linefmt='r--', markerfmt='ro', basefmt='r-', label='Output Signal')
plt.title('Difference Equation')
plt.xlabel('n (Sample Index)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a discrete signal with two frequencies
N = 64
n = np.arange(N)
x = np.sin(2 * np.pi * 5 * n / N) + 0.5 * np.sin(2 * np.pi * 12 * n / N)

# Compute the DFT
X = np.fft.fft(x)
frequencies = np.fft.fftfreq(N)

# Plot the magnitude spectrum
plt.stem(frequencies, np.abs(X))
plt.title('DFT - Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a noisy signal
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(1000)

# Apply a moving average filter
window_size = 10
moving_avg = np.convolve(signal, np.ones(window_size) / window_size, 'valid')

# Plot the original and filtered signals
plt.plot(t, signal, label='Original Signal')
plt.plot(t[window_size - 1:], moving_avg, label='Moving Average Filtered Signal')
plt.title('Signal Processing with Moving Average Filter')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

# Compute 2D Fourier Transform
f_transform = np.fft.fft2(image)
f_transform_shifted = np.fft.fftshift(f_transform)

# Calculate the magnitude spectrum
magnitude_spectrum = np.log(np.abs(f_transform_shifted) + 1)

# Display the original and magnitude spectrum
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.axis('off')
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum'), plt.axis('off')
plt.show()
```

```python
import librosa
import matplotlib.pyplot as plt

# Load an audio file
audio_file = "sample.wav"
y, sr = librosa.load(audio_file, sr=None)

# Extract formants using LPC analysis
order = 12  # Order of LPC analysis
formants = librosa.effects.lpc(y, order)

# Plot the LPC coefficients (formants)
plt.plot(formants)
plt.title('LPC Coefficients (Formants)')
plt.xlabel('Coefficient Index')
plt.ylabel('Magnitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl

# Define a transfer function for a simple control system
numerator = [1]
denominator = [1, 2, 1]
system = ctl.TransferFunction(numerator, denominator)

# Time vector
t = np.linspace(0, 10, 1000)

# Simulate the step response
t, y = ctl.step_response(system, T=t)

# Plot the step response
plt.plot(t, y)
plt.title('Step Response of the Control System')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
