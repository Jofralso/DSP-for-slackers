Make a square signal at 1100 Hz and make a wave that samples it at 10000 frames per second. If you plot the spectrum, you can see that most of the harmonics are aliased. When you listen to the wave, can you hear the aliased harmonics?

```python
import os
import urllib.request

file_url = "https://github.com/AllenDowney/ThinkDSP/raw/master/code/thinkdsp.py"
destination_file = "thinkdsp.py"

if not os.path.exists(destination_file):
    urllib.request.urlretrieve(file_url, destination_file)
    print("File downloaded successfully.")
else:
    print("File already exists.")
```

```python

from thinkdsp import SquareSignal, decorate, read_wave

square =SquareSignal(1100).make_wave(duration=0.5, framerate=22000)
```

```python
sq_spec = square.make_spectrum()
sq_spec.plot()
decorate(xlabel='Freq Hz')
```

```python
sq_spec.peaks()[:4]
```
